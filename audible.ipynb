```python
import pandas as pd
```

```python
file_path = '/home/jakubg/Desktop/csv/svms.csv'
data = pd.read_csv(file_path, sep='\t', encoding="utf-16")
data
```

```python
result = data[data['Polish_Text'].notna()]

result
```

```python
import os
start = "/home/jakubg/Desktop/speech_english/SVM"
search = "/home/jakubg/Desktop/test"

audible = []
for index, row in data.iterrows():
    polish_text = row['Polish_Text']
    file_name = row['Key']
    if pd.notna(polish_text):
        string = polish_text + "{-}" + file_name
        audible.append(string)

for root, dirs, files in os.walk(search):
    if root == search:
        for dir in dirs:
            concat = os.path.join(search,dir,"audible")
            for file_name in audible:
                if dir in file_name:
                    parts = file_name.split('{-}',1)
                    part1, part2 = parts
                    file_path = os.path.join(concat,part2)
                    with open(file_path, "w") as file:
                        file.write(part1)
                        
```
