```python




def infos(start_path,end_path):

    if not os.path.exists(end_path):
        os.makedirs(end_path,exist_ok=True)

    
    for directory in os.listdir(start_path): 
        os.makedirs(os.path.join(end_path,directory),exist_ok=True)
        path = os.path.join(start_path, directory)

        for object in os.listdir(path):
            obj_path = os.path.join(path,object)
            if os.path.isdir(obj_path):
                dir_path = os.path.join(end_path, directory, object)
                os.makedirs(dir_path,exist_ok=True)
                
                for file in os.listdir(obj_path):
                    if file.endswith('xsnd'):
                        
                        index = file.find("._xsnd")
    
                        new_file = file[:index]
                        part = file.split('_')
                        table_name = '_'.join(part[:3])

                        result = data_subset[data_subset['ID'] == table_name]
                        if not result.empty and result['Polish_Text'].notna().any():
                            polish_text_value = result['Polish_Text'].iloc[0]

                            new_path = os.path.join(dir_path,new_file)
                            with open(new_path,'w') as file:
                                file.write(polish_text_value)

            # PLIKI W END_PATH
            else:
                if object.endswith('xsnd'):

                    index = object.find("._xsnd")

                    new_file = object[:index]
                    part = object.split('_')
                    table_name = '_'.join(part[:3])
                    result = data_subset[data_subset['ID'] == table_name]
                    if not result.empty and result['Polish_Text'].notna().any():
                        polish_text_value = result['Polish_Text'].iloc[0]
                        new_path = os.path.join(end_path,directory,new_file)
                        with open(new_path,'w') as file:
                            file.write(polish_text_value)
start ='/home/jakubg/Desktop/speech_english/infos' 
end ='/home/jakubg/Desktop/test/infos'

infos(start,end)
```

```python
import os
import subprocess
import shutil

start = '/home/jakubg/PACKAGE/test/kopia_out'
end = '/home/jakubg/PACKAGE/info'

def transfer(source,destination):
    os.makedirs(end,exist_ok=True)
    for directory in os.listdir(source):
        dir_path = os.path.join(source,directory)
        out_dir_path = os.path.join(destination,directory)
        
        os.makedirs(out_dir_path, exist_ok=True)
        for object in os.listdir(dir_path):
            object_path = os.path.join(source,directory,object)
            out_path = os.path.join(out_dir_path,object)
            if os.path.isfile(object_path):
                if 'xsnd' in object:
                    shutil.move(object_path,out_path)
            else:
                os.makedirs(out_path,exist_ok=True)
                for file in os.listdir(object_path):
                    if 'xsnd' in file:
                        object_path = os.path.join(source,directory,object,file)
                        out_path = os.path.join(destination,directory,object,file)
                        shutil.move(object_path,out_path)
transfer(start,end)               
```
