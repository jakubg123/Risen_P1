```python
import plotly.express as px
def draw(dict, save):
    names = list(dict.keys())
    avg_volume = [db['average_volume_dBFS'] for db in dict.values()]
    
    data = {'Character' : names, 'Avg volume' : avg_volume}
    
    
    fig = px.bar(data, y='Character', x='Avg volume')
    fig.write_html(save)
```

```python
def boost_audio(file_path, boost_amount,threshold):
    try:
        audio = AudioSegment.from_mp3(file_path)
        current_dBFS = audio.dBFS
        if current_dBFS > threshold and current_dBFS < -12:
            audio+=3
        elif current_dBFS < threshold:
            audio+=8
            
        audio.export(file_path, format="mp3")
        print(f"{file_path}")
    except Exception as e:
        print({e})
        
threshold = -14
boost_amount = 10
for char_data in graph.values():
    file_paths = char_data['file_paths']
    
    for file_path in file_paths:
        boost_audio(file_path, boost_amount,threshold)

```
